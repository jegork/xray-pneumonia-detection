# A small exploratory analysis of the dataset

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
plt.style.use('seaborn')
```

```python
train_normal_size = len(os.listdir('chest_xray/train/NORMAL'))
train_pneumonia_size = len(os.listdir('chest_xray/train/PNEUMONIA'))

test_normal_size = len(os.listdir('chest_xray/test/NORMAL'))
test_pneumonia_size = len(os.listdir('chest_xray/test/PNEUMONIA'))

val_normal_size = len(os.listdir('chest_xray/val/NORMAL'))
val_pneumonia_size = len(os.listdir('chest_xray/val/PNEUMONIA'))
```

```python
datasets_sizes = [
            train_normal_size + train_pneumonia_size, 
            test_normal_size + test_pneumonia_size,
            val_normal_size + val_pneumonia_size
        ]

plt.bar(['Train', 'Test', 'Validation'], 
        datasets_sizes)

for i, v in enumerate(datasets_sizes):
    plt.text(i, v + 50, v, c='black', fontweight='bold', ha='center')

plt.title('Number of images in each data set')
plt.ylabel('Count')

plt.show()
```

### Let's see how many images of each class we have

```python
fig, ax = plt.subplots(ncols=3, figsize=(10, 4))

fig.suptitle('Distribution of classes in each dataset', y=1.03)

ax[0].bar(['Normal', 'Pneumonia'], [train_normal_size, train_pneumonia_size])
ax[0].set_ylabel('Count')
ax[0].set_title('Training')

ax[1].bar(['Normal', 'Pneumonia'], [test_normal_size, test_pneumonia_size])
ax[1].set_title('Test')

ax[2].bar(['Normal', 'Pneumonia'], [val_normal_size, val_pneumonia_size])
ax[2].set_title('Validation')

fig.show()
```

We can see that the 'pneumonia' class is much more frequent in the training dataset

```python
fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(12, 15))

for i, cls in enumerate(['NORMAL', 'PNEUMONIA']):
    path = os.path.join('chest_xray/train', cls)
    imgs = os.listdir(path)
    
    ax[0][i].set_title(cls)
    
    for row in range(3):
        img = mpimg.imread(os.path.join(path, imgs[row]))
        ax[row][i].imshow(img, aspect='auto', cmap='Greys_r')
        ax[row][i].grid(False)

fig.suptitle('Examples of images of each class', y=0.92)
fig.show()
```
